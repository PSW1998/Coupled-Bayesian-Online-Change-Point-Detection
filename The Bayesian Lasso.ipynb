Assume for simplicity that we have demeaned all the variables so that $y_{i}= y_{i}-\bar{y}$ and $x_{ij} = x_{ij}-\bar{x}_{.j}$.

$$y|\beta,\sigma^2\sim N(X\beta,\sigma^2I_{n})$$

$$\beta_{j}|\tau_{j}^2,\sigma^2 \sim N(0,\sigma^2\tau_{j}^2), \quad \tau_{j}^2|\lambda^2\sim Exp(\lambda^2/2) \Rightarrow \beta_{j}|\sigma^2,\lambda \sim Laplace (0,\sigma/\lambda)$$

$$\lambda^2\sim G(r,s), \quad \sigma^2 \sim IG(a_{0},b_{0})$$

**Posterior Distribution for $\beta$.**

$$p(\beta|\tilde{y},\tau_{j}^2,\sigma^2)\propto p(\beta|\tau^2,\sigma^2)p(y|\beta,\sigma^2)\propto \exp(-\frac{1}{2\sigma^2}\beta^{\top}D_{\tau}^{-1}\beta)\exp(-\frac{1}{2\sigma^2}(y-X\beta)^{\top}(y-X\beta))$$

$$\propto \exp(-\frac{1}{2\sigma^2}(\beta^{\top}D_{\tau}^{-1}\beta+(y-X\beta)^{\top}(y-X\beta)))\propto \exp(-\frac{1}{2\sigma^2}(\beta^{\top}(X^{\top}X+D_{\tau}^{-1})\beta+2\beta^{\top}X^{\top}\tilde{y}))\sim N((X^{\top}X+D_{\tau}^{-1})^{-1}X^{\top}\tilde{y},(X^{\top}X+D_{\tau}^{-1})^{-1})$$

**Posterior Distribution for $\tau^2$.**

$$
p(\tau^2 \mid \beta,\sigma^2,\lambda^2)
\propto
p(\beta \mid \tau^2,\sigma^2)\,p(\tau^2 \mid \lambda^2).
\propto
p(\tau^2 \mid \beta,\sigma^2,\lambda^2)
\;\propto\;
(\tau^2)^{-1/2}
\exp\!\left(
-\frac{\beta^2}{2\sigma^2\tau^2}
-\frac{\lambda^2}{2}\tau^2
\right)
$$


$$
z = \frac{1}{\tau^2}, \qquad \tau^2 = \frac{1}{z}, \qquad 
\left|\frac{d\tau^2}{dz}\right| = \frac{1}{z^2}.
$$



\begin{aligned}
p(z \mid \beta,\sigma^2,\lambda^2)
&= p(\tau^2 = 1/z \mid \beta,\sigma^2,\lambda^2)\,\frac{1}{z^2} \\
&\propto
\left(\frac{1}{z}\right)^{-1/2}
\exp\!\left(
-\frac{\beta^2}{2\sigma^2/(1/z)}
-\frac{\lambda^2}{2}\cdot\frac{1}{z}
\right)\frac{1}{z^2} \\
&=
z^{1/2}
\exp\!\left(
-\frac{\beta^2 z}{2\sigma^2}
-\frac{\lambda^2}{2z}
\right) z^{-2} \\
&=
z^{-3/2}
\exp\!\left(
-\frac{\beta^2}{2\sigma^2} z
-\frac{\lambda^2}{2z}
\right),
\qquad z>0.
\end{aligned}

$$
p(z \mid \beta,\sigma^2,\lambda^2)
\;\propto\;
z^{-3/2}
\exp\!\left(
-\frac{\beta^2}{2\sigma^2} z
-\frac{\lambda^2}{2z}
\right).
$$

$$
z = \frac{1}{\tau^2} \,\bigg|\, \beta,\sigma^2,\lambda^2
\sim \mathrm{IG}\!\left(
\mu = \frac{\lambda\sigma}{|\beta|},\;
\lambda_{\mathrm{IG}} = \lambda^2
\right).
$$


**Posterior Distribution for $\lambda$.**

$$
p(\lambda^2 \mid \tau_1^2,\dots,\tau_p^2)
\propto
p(\lambda^2)\, \prod_{j=1}^p p(\tau_j^2 \mid \lambda^2).
$$


\begin{aligned}
p(\lambda^2 \mid \tau_1^2,\dots,\tau_p^2)
&\propto
(\lambda^2)^{r-1} e^{-s\lambda^2}
\cdot
(\lambda^2)^p
\exp\!\left(-\frac{\lambda^2}{2}\sum_{j=1}^p \tau_j^2\right) \\
&=
(\lambda^2)^{r-1+p}
\exp\!\left\{
-\left(
s + \frac{1}{2}\sum_{j=1}^p \tau_j^2
\right)\lambda^2
\right\}.
\end{aligned}



$$
\lambda^2 \mid \tau_1^2,\dots,\tau_p^2
\sim
\mathrm{Gamma}\left(
r + p,\;
s + \frac{1}{2}\sum_{j=1}^p \tau_j^2
\right).
$$


**Posterior Distribution for $\sigma^2$.**

\begin{aligned}
p(\sigma^2 \mid y,\beta,\tau^2)
&\propto
p(y \mid \beta,\sigma^2)\,
p(\beta \mid \tau^2,\sigma^2)\,
p(\sigma^2) \\
&\propto
(\sigma^2)^{-n/2}
\exp\!\left(-\frac{1}{2\sigma^2}\|r\|^2\right)
\cdot
(\sigma^2)^{-p/2}
\exp\!\left(-\frac{1}{2\sigma^2}\beta^\top D_\tau^{-1}\beta\right)
\\
&\qquad \cdot
(\sigma^2)^{-(a_0+1)}
\exp\!\left(-\frac{b_0}{\sigma^2}\right).
\end{aligned}

$$
p(\sigma^2 \mid y,\beta,\tau^2)
\propto
(\sigma^2)^{-\left(a_0 + \frac{n+p}{2} + 1\right)}
\exp\!\left\{
-\frac{1}{\sigma^2}
\left[
b_0 + \frac{1}{2}\|y - X\beta\|^2
      + \frac{1}{2}\sum_{j=1}^p \frac{\beta_j^2}{\tau_j^2}
\right]
\right\}.
$$

$$
\sigma^2 \mid y,\beta,\tau^2
\sim
\mathrm{IG}\!\left(
a_0 + \frac{n+p}{2},\;
b_0 + \frac{1}{2}\|y - X\beta\|^2
     + \frac{1}{2}\sum_{j=1}^p \frac{\beta_j^2}{\tau_j^2}
\right).
$$


```python
import numpy as np

# ---------- 1. Inverse-Gaussian sampler ----------

def sample_inverse_gaussian(mu, lam):
    """
    Sample from IG(mu, lam) using the Michael–Schucany–Haas method.
    mu > 0, lam > 0.
    """
    v = np.random.normal()
    y = v**2
    x = mu + (mu**2 * y) / (2 * lam) - (mu / (2 * lam)) * np.sqrt(4 * mu * lam * y + mu**2 * y**2)
    u = np.random.rand()
    if u <= mu / (mu + x):
        return x
    else:
        return mu**2 / x

# ---------- 2. Bayesian lasso Gibbs sampler ----------

def bayesian_lasso_gibbs(X, y, a0=1.0, b0=1.0, r=1.0, s=1.0,
                         T=4000, burn_in=2000, random_state=None):
    """
    Centered model: y | beta, sigma^2 ~ N(X beta, sigma^2 I).
    Fully Bayesian lasso with:
      beta_j | tau_j^2, sigma^2 ~ N(0, sigma^2 tau_j^2)
      tau_j^2 | lambda^2 ~ Exp(lambda^2 / 2)
      lambda^2 ~ Gamma(r, s)        (shape-rate)
      sigma^2 ~ Inv-Gamma(a0, b0)   (shape-scale)
    Returns posterior samples of beta after burn-in.
    """
    if random_state is not None:
        np.random.seed(random_state)

    n, p = X.shape

    # Initialize
    beta = np.zeros(p)
    tau2 = np.ones(p)
    lam2 = 1.0
    sigma2 = 1.0

    betas = []

    XtX = X.T @ X
    Xty = X.T @ y

    for t in range(T):
        # 1. Update beta | y, tau^2, sigma^2
        Dtau_inv = np.diag(1.0 / tau2)
        Vbeta_inv = XtX + Dtau_inv
        Vbeta = np.linalg.inv(Vbeta_inv)
        mbeta = Vbeta @ Xty
        beta = np.random.multivariate_normal(mbeta, sigma2 * Vbeta)

        # 2. Update tau_j^2 via z_j = 1 / tau_j^2
        for j in range(p):
            bj = beta[j]
            if bj == 0:
                bj = 1e-8  # avoid division by zero
            mu_j = np.sqrt((lam2 * sigma2) / (bj**2))  # = (lambda * sigma) / |beta_j|
            z_j = sample_inverse_gaussian(mu_j, lam2)  # z_j ~ IG(mu_j, lam2)
            tau2[j] = 1.0 / z_j

        # 3. Update lambda^2 | tau^2
        shape_lam = r + p
        rate_lam = s + 0.5 * tau2.sum()
        lam2 = np.random.gamma(shape_lam, 1.0 / rate_lam)  # Gamma(shape, scale=1/rate)

        # 4. Update sigma^2 | y, beta, tau^2
        resid = y - X @ beta
        SSE = resid @ resid
        quad = np.sum(beta**2 / tau2)

        shape_sig = a0 + 0.5 * (n + p)
        scale_sig = b0 + 0.5 * (SSE + quad)  # IG(a,b): density ∝ x^{-(a+1)} exp(-b/x)
        sigma2 = 1.0 / np.random.gamma(shape_sig, 1.0 / scale_sig)

        # Store after burn-in
        if t >= burn_in:
            betas.append(beta.copy())

    return np.array(betas)

# ---------- 3. Simulate data and test ----------

if __name__ == "__main__":
    np.random.seed(42)

    n, p = 1500, 10
    X_raw = np.random.normal(size=(n, p))

    # Center columns of X
    X = X_raw - X_raw.mean(axis=0, keepdims=True)

    # True sparse beta
    beta_true = np.array([3.0, -2.0, 0.0, 1.5] + [0.0] * (p - 4))
    sigma_true = 1.0

    # Generate y and center
    y_raw = X @ beta_true + np.random.normal(scale=sigma_true, size=n)
    y = y_raw - y_raw.mean()

    # Run Gibbs sampler
    betas_samples = bayesian_lasso_gibbs(X, y, a0=1.0, b0=1.0,
                                         r=1.0, s=1.0,
                                         T=4000, burn_in=2000,
                                         random_state=123)

    beta_post_mean = betas_samples.mean(axis=0)
    beta_post_sd = betas_samples.std(axis=0)

    print("True beta:      ", beta_true)
    print("Post mean beta: ", np.round(beta_post_mean, 3))
    print("Post sd beta:   ", np.round(beta_post_sd, 3))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# ============================================================
# 1) Inverse-Gaussian sampler  IG(mu, lam)
# ============================================================

def sample_inverse_gaussian(mu, lam):
    """
    Sample from an Inverse-Gaussian IG(mu, lam) using the
    Michael–Schucany–Haas method.

    Parameters
    ----------
    mu : float
        Mean parameter (> 0).
    lam : float
        Shape parameter (> 0).

    Returns
    -------
    float
        A single draw from IG(mu, lam).
    """
    if mu <= 0 or lam <= 0:
        raise ValueError("mu and lam must be > 0")

    v = np.random.normal()
    y = v**2
    mu2 = mu**2

    x = mu + (mu2 * y) / (2 * lam) - (mu / (2 * lam)) * np.sqrt(
        4 * mu * lam * y + mu2 * y**2
    )
    u = np.random.rand()
    if u <= mu / (mu + x):
        return x
    else:
        return mu2 / x


# ============================================================
# 2) Bayesian lasso Gibbs sampler (centered model)
# ============================================================

def bayesian_lasso_gibbs(
    X,
    y,
    T=5000,
    burn_in=2000,
    a0=1.0,
    b0=1.0,
    r=1.0,
    s=1.0,
    random_state=None,
):
    """
    Fully Bayesian lasso via Gibbs sampling.

    Centered model (no intercept):
        y | beta, sigma^2 ~ N(X beta, sigma^2 I_n)

        beta_j | tau_j^2, sigma^2 ~ N(0, sigma^2 tau_j^2)
        tau_j^2 | lambda^2 ~ Exp(lambda^2 / 2)

        lambda^2 ~ Gamma(r, s)       (shape-rate)
        sigma^2  ~ Inv-Gamma(a0, b0) (shape-scale)

    Assumes y and columns of X are already centered.

    Returns dict of posterior draws for beta, tau^2, lambda^2, sigma^2.
    """
    if random_state is not None:
        np.random.seed(random_state)

    n, p = X.shape

    # Initialize parameters
    beta = np.zeros(p)
    tau2 = np.ones(p)
    lam2 = 1.0
    sigma2 = 1.0

    num_save = T - burn_in
    beta_samples = np.zeros((num_save, p))
    tau2_samples = np.zeros((num_save, p))
    lambda2_samples = np.zeros(num_save)
    sigma2_samples = np.zeros(num_save)

    # Precompute for speed
    XtX = X.T @ X
    Xty = X.T @ y

    for t in range(T):
        # ----------------------------------------------------
        # 1) beta | y, tau2, sigma2  ~ N(m_beta, sigma2 V_beta)
        # ----------------------------------------------------
        Dtau_inv = np.diag(1.0 / tau2)
        Vbeta_inv = XtX + Dtau_inv
        Vbeta = np.linalg.inv(Vbeta_inv)
        mbeta = Vbeta @ Xty

        # multivariate normal via Cholesky
        L = np.linalg.cholesky(sigma2 * Vbeta)
        z = np.random.normal(size=p)
        beta = mbeta + L @ z

        # ----------------------------------------------------
        # 2) tau_j^2 | beta_j, sigma2, lambda2  (via z_j = 1/tau_j^2)
        # ----------------------------------------------------
        for j in range(p):
            bj = beta[j]
            if bj == 0.0:
                bj = 1e-8  # avoid division by zero
            # mu_j = sqrt( (lambda^2 * sigma^2) / beta_j^2 )
            mu_j = np.sqrt((lam2 * sigma2) / (bj**2))
            z_j = sample_inverse_gaussian(mu_j, lam2)
            tau2[j] = 1.0 / z_j

        # ----------------------------------------------------
        # 3) lambda^2 | tau2  ~ Gamma(r + p, s + 0.5 * sum tau2)
        # ----------------------------------------------------
        shape_lam = r + p
        rate_lam = s + 0.5 * tau2.sum()
        lam2 = np.random.gamma(shape_lam, 1.0 / rate_lam)

        # ----------------------------------------------------
        # 4) sigma^2 | y, beta, tau2  ~ Inv-Gamma(...)
        #     sample as 1 / Gamma(a*, 1/b*)
        # ----------------------------------------------------
        resid = y - X @ beta
        SSE = resid @ resid
        quad = np.sum(beta**2 / tau2)

        shape_sig = a0 + 0.5 * (n + p)
        scale_sig = b0 + 0.5 * (SSE + quad)
        sigma2 = 1.0 / np.random.gamma(shape_sig, 1.0 / scale_sig)

        # ----------------------------------------------------
        # Save draws after burn-in
        # ----------------------------------------------------
        if t >= burn_in:
            idx = t - burn_in
            beta_samples[idx, :] = beta
            tau2_samples[idx, :] = tau2
            lambda2_samples[idx] = lam2
            sigma2_samples[idx] = sigma2

    return {
        "beta": beta_samples,
        "tau2": tau2_samples,
        "lambda2": lambda2_samples,
        "sigma2": sigma2_samples,
    }


# ============================================================
# 3) Simulate data
# ============================================================

np.random.seed(42)

n, p = 150, 10
X_raw = np.random.normal(size=(n, p))

# Center X columns
X = X_raw - X_raw.mean(axis=0, keepdims=True)

# True sparse coefficients
beta_true = np.array([3.0, -2.0, 0.0, 1.5] + [0.0] * (p - 4))
sigma_true = 1.0

# Generate y and center
y_raw = X @ beta_true + np.random.normal(scale=sigma_true, size=n)
y = y_raw - y_raw.mean()

# ============================================================
# 4) Run Bayesian lasso Gibbs sampler
# ============================================================

draws = bayesian_lasso_gibbs(
    X,
    y,
    T=5000,
    burn_in=2000,
    a0=1.0,
    b0=1.0,
    r=1.0,
    s=1.0,
    random_state=123,
)

beta_samples = draws["beta"]
tau2_samples = draws["tau2"]
lambda2_samples = draws["lambda2"]
sigma2_samples = draws["sigma2"]

beta_post_mean = beta_samples.mean(axis=0)

print("True beta:      ", beta_true)
print("Post mean beta: ", np.round(beta_post_mean, 3))


# ============================================================
# 5) Plot posterior distributions for beta, tau^2, lambda^2, sigma^2
# ============================================================

# ---- beta_j posteriors ----
for j in range(p):
    plt.figure()
    plt.hist(beta_samples[:, j], bins=40, density=True)
    plt.axvline(beta_true[j], linestyle='--')
    plt.title(f"Posterior of beta[{j+1}]")
    plt.xlabel("beta value")
    plt.ylabel("Density")
    plt.tight_layout()
    plt.show()

# ---- tau_j^2 posteriors ----
for j in range(p):
    plt.figure()
    plt.hist(tau2_samples[:, j], bins=40, density=True)
    plt.title(f"Posterior of tau^2[{j+1}]")
    plt.xlabel("tau^2 value")
    plt.ylabel("Density")
    plt.tight_layout()
    plt.show()

# ---- lambda^2 posterior ----
plt.figure()
plt.hist(lambda2_samples, bins=40, density=True)
plt.xlabel(r"$\lambda^2$")
plt.ylabel("Density")
plt.title(r"Posterior of $\lambda^2$")
plt.tight_layout()
plt.show()

# ---- sigma^2 posterior ----
plt.figure()
plt.hist(sigma2_samples, bins=40, density=True)
plt.axvline(sigma_true, linestyle='--')  # true sigma^2 = 1
plt.xlabel(r"$\sigma^2$")
plt.ylabel("Density")
plt.title(r"Posterior of $\sigma^2$")
plt.tight_layout()
plt.show()


# ============================================================
# 6) Posterior predictive vs true predictive at one x_0
# ============================================================

# Choose one covariate vector x0 (from centered X)
x0 = X[0, :]
true_mean = x0 @ beta_true
true_sd = sigma_true

# Posterior predictive draws: y0^(s) = x0^T beta^(s) + eps^(s)
S = beta_samples.shape[0]
linpred_samples = beta_samples @ x0  # shape (S,)
# draw one epsilon per posterior sample
eps = np.random.normal(scale=np.sqrt(sigma2_samples))
y_pred_samples = linpred_samples + eps

# Plot posterior predictive histogram and true predictive density
plt.figure()
plt.hist(y_pred_samples, bins=40, density=True, alpha=0.6, label="Posterior predictive")

# Overlay true predictive N(x0^T beta_true, sigma_true^2)
xs = np.linspace(y_pred_samples.min(), y_pred_samples.max(), 200)
plt.plot(xs, norm.pdf(xs, loc=true_mean, scale=true_sd), linestyle='--', label="True predictive")

plt.xlabel("y at x0")
plt.ylabel("Density")
plt.title("Posterior predictive vs true predictive at x0")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from scipy.stats import norm

# ---------- 1. Load and center real data (diabetes) ----------

data = load_diabetes()
X_raw = data.data      # shape (n, p), already standardized-ish
y_raw = data.target    # disease progression

n, p = X_raw.shape
print("n, p =", n, p)

# Center columns of X and y (no intercept in the model)
X = X_raw - X_raw.mean(axis=0, keepdims=True)
y = y_raw - y_raw.mean()

# ---------- 2. Run Bayesian lasso Gibbs sampler ----------

draws = bayesian_lasso_gibbs(
    X,
    y,
    T=6000,        # total iterations
    burn_in=3000,  # burn-in
    a0=1.0,
    b0=1.0,
    r=1.0,
    s=1.0,
    random_state=123,
)

beta_samples   = draws["beta"]     # (S, p)
tau2_samples   = draws["tau2"]     # (S, p)
lambda2_samples = draws["lambda2"] # (S,)
sigma2_samples = draws["sigma2"]   # (S,)

beta_post_mean = beta_samples.mean(axis=0)
beta_post_sd   = beta_samples.std(axis=0)

print("Posterior mean beta:", np.round(beta_post_mean, 3))
print("Posterior sd beta:  ", np.round(beta_post_sd, 3))

# ---------- 3. Plot posterior distributions for beta ----------

for j in range(p):
    plt.figure()
    plt.hist(beta_samples[:, j], bins=40, density=True)
    plt.axvline(beta_post_mean[j], linestyle="--", label="Posterior mean")
    plt.title(f"Posterior of beta[{j}]")
    plt.xlabel("beta value")
    plt.ylabel("Density")
    plt.legend()
    plt.tight_layout()
    plt.show()

# ---------- 4. Plot posterior for lambda^2 and sigma^2 ----------

plt.figure()
plt.hist(lambda2_samples, bins=40, density=True)
plt.xlabel(r"$\lambda^2$")
plt.ylabel("Density")
plt.title(r"Posterior of $\lambda^2$")
plt.tight_layout()
plt.show()

plt.figure()
plt.hist(sigma2_samples, bins=40, density=True)
plt.xlabel(r"$\sigma^2$")
plt.ylabel("Density")
plt.title(r"Posterior of $\sigma^2$")
plt.tight_layout()
plt.show()

# ---------- 5. Posterior predictive vs "real" predictive ----------

# Pick some x0 from the dataset (centered version)
x0 = X[0, :]     # 1D array of length p

# Posterior predictive draws: y0^(s) = x0^T beta^(s) + eps^(s)
S = beta_samples.shape[0]
linpred_samples = beta_samples @ x0     # shape (S,)
eps = np.random.normal(scale=np.sqrt(sigma2_samples))  # one noise draw per sample
y_pred_samples = linpred_samples + eps

# For a "true" reference, we don't know the real generative model on diabetes,
# but we can approximate with:
mu_hat = linpred_samples.mean()
sd_hat = np.sqrt(sigma2_samples.mean())

plt.figure()
plt.hist(y_pred_samples, bins=40, density=True, alpha=0.6, label="Posterior predictive")

xs = np.linspace(y_pred_samples.min(), y_pred_samples.max(), 200)
plt.plot(xs, norm.pdf(xs, loc=mu_hat, scale=sd_hat),
         linestyle="--", label="Gaussian approx (mean+var)")

plt.xlabel("y at x0")
plt.ylabel("Density")
plt.title("Posterior predictive at one x0 (diabetes data)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm

# ============================================================
# 1) Inverse-Gaussian sampler  IG(mu, lam)
# ============================================================

def sample_inverse_gaussian(mu, lam):
    """
    Sample from an Inverse-Gaussian IG(mu, lam) using
    the Michael–Schucany–Haas method.

    Parameters
    ----------
    mu : float
        Mean parameter (> 0).
    lam : float
        Shape parameter (> 0).

    Returns
    -------
    float
        A single draw from IG(mu, lam).
    """
    if mu <= 0 or lam <= 0:
        raise ValueError("mu and lam must be > 0")

    v = np.random.normal()
    y = v**2
    mu2 = mu**2

    x = mu + (mu2 * y) / (2 * lam) - (mu / (2 * lam)) * np.sqrt(
        4 * mu * lam * y + mu2 * y**2
    )
    u = np.random.rand()
    if u <= mu / (mu + x):
        return x
    else:
        return mu2 / x


# ============================================================
# 2) Bayesian lasso Gibbs sampler (centered model)
# ============================================================

def bayesian_lasso_gibbs(
    X,
    y,
    T=6000,
    burn_in=3000,
    a0=1.0,
    b0=1.0,
    r=1.0,
    s=1.0,
    random_state=None,
):
    """
    Fully Bayesian lasso via Gibbs sampling.

    Centered model (no intercept):
        y | beta, sigma^2 ~ N(X beta, sigma^2 I_n)

        beta_j | tau_j^2, sigma^2 ~ N(0, sigma^2 tau_j^2)
        tau_j^2 | lambda^2 ~ Exp(lambda^2 / 2)

        lambda^2 ~ Gamma(r, s)       (shape-rate)
        sigma^2  ~ Inv-Gamma(a0, b0) (shape-scale)

    Assumes y and columns of X are already centered.

    Returns dict of posterior draws for beta, tau^2, lambda^2, sigma^2.
    """
    if random_state is not None:
        np.random.seed(random_state)

    n, p = X.shape

    # Initialize parameters
    beta = np.zeros(p)
    tau2 = np.ones(p)
    lam2 = 1.0
    sigma2 = 1.0

    num_save = T - burn_in
    beta_samples = np.zeros((num_save, p))
    tau2_samples = np.zeros((num_save, p))
    lambda2_samples = np.zeros(num_save)
    sigma2_samples = np.zeros(num_save)

    # Precompute for speed
    XtX = X.T @ X
    Xty = X.T @ y

    for t in range(T):
        # ----------------------------------------------------
        # 1) beta | y, tau2, sigma2  ~ N(m_beta, sigma2 V_beta)
        # ----------------------------------------------------
        Dtau_inv = np.diag(1.0 / tau2)
        Vbeta_inv = XtX + Dtau_inv
        Vbeta = np.linalg.inv(Vbeta_inv)
        mbeta = Vbeta @ Xty

        # multivariate normal via Cholesky
        L = np.linalg.cholesky(sigma2 * Vbeta)
        z = np.random.normal(size=p)
        beta = mbeta + L @ z

        # ----------------------------------------------------
        # 2) tau_j^2 | beta_j, sigma2, lambda2  (via z_j = 1/tau_j^2)
        # ----------------------------------------------------
        for j in range(p):
            bj = beta[j]
            if bj == 0.0:
                bj = 1e-8  # avoid division by zero
            # mu_j = sqrt( (lambda^2 * sigma^2) / beta_j^2 )
            mu_j = np.sqrt((lam2 * sigma2) / (bj**2))
            z_j = sample_inverse_gaussian(mu_j, lam2)
            tau2[j] = 1.0 / z_j

        # ----------------------------------------------------
        # 3) lambda^2 | tau2  ~ Gamma(r + p, s + 0.5 * sum tau2)
        # ----------------------------------------------------
        shape_lam = r + p
        rate_lam = s + 0.5 * tau2.sum()
        lam2 = np.random.gamma(shape_lam, 1.0 / rate_lam)

        # ----------------------------------------------------
        # 4) sigma^2 | y, beta, tau2  ~ Inv-Gamma(...)
        #     sample as 1 / Gamma(a*, 1/b*)
        # ----------------------------------------------------
        resid = y - X @ beta
        SSE = resid @ resid
        quad = np.sum(beta**2 / tau2)

        shape_sig = a0 + 0.5 * (n + p)
        scale_sig = b0 + 0.5 * (SSE + quad)
        sigma2 = 1.0 / np.random.gamma(shape_sig, 1.0 / scale_sig)

        # ----------------------------------------------------
        # Save draws after burn-in
        # ----------------------------------------------------
        if t >= burn_in:
            idx = t - burn_in
            beta_samples[idx, :] = beta
            tau2_samples[idx, :] = tau2
            lambda2_samples[idx] = lam2
            sigma2_samples[idx] = sigma2

    return {
        "beta": beta_samples,
        "tau2": tau2_samples,
        "lambda2": lambda2_samples,
        "sigma2": sigma2_samples,
    }


# ============================================================
# 3) Load and preprocess ISLR "Hitters" data
# ============================================================

# Raw CSV from selva86/datasets (ISLR data collection)
url = "https://raw.githubusercontent.com/selva86/datasets/master/Hitters.csv"
hitters = pd.read_csv(url)

# Drop missing Salary, as in ISLR
hitters = hitters.dropna(subset=["Salary"])

# Log-transform Salary (ISLR uses log-salary)
y_raw = np.log(hitters["Salary"].values)

# Drop Salary and name column, keep numeric + categorical predictors
X_df = hitters.drop(columns=["Salary"])

# Some versions have an "Unnamed: 0" or player name column; drop if present
for col in ["Unnamed: 0", "Player"]:
    if col in X_df.columns:
        X_df = X_df.drop(columns=[col])

# One-hot encode categorical vars (League, Division, NewLeague) with drop_first
X_df = pd.get_dummies(X_df, drop_first=True)

# Convert to numpy and center
X = X_df.values.astype(float)
X = X - X.mean(axis=0, keepdims=True)
y = y_raw - y_raw.mean()

n, p = X.shape
print(f"Hitters: n={n}, p={p} (after encoding & centering)")


# ============================================================
# 4) Run Bayesian lasso Gibbs sampler
# ============================================================

draws = bayesian_lasso_gibbs(
    X,
    y,
    T=6000,
    burn_in=3000,
    a0=1.0,
    b0=1.0,
    r=1.0,
    s=1.0,
    random_state=123,
)

beta_samples   = draws["beta"]      # (S, p)
tau2_samples   = draws["tau2"]      # (S, p)
lambda2_samples = draws["lambda2"]  # (S,)
sigma2_samples = draws["sigma2"]    # (S,)

beta_post_mean = beta_samples.mean(axis=0)
beta_post_sd   = beta_samples.std(axis=0)

print("Posterior mean beta (first 10):", np.round(beta_post_mean[:10], 3))
print("Posterior sd beta   (first 10):", np.round(beta_post_sd[:10], 3))


# ============================================================
# 5) Plot posterior distributions
# ============================================================

# ---- beta_j posteriors ----
feature_names = X_df.columns.tolist()

for j in range(p):
    plt.figure()
    plt.hist(beta_samples[:, j], bins=40, density=True)
    plt.axvline(beta_post_mean[j], linestyle="--", label="Posterior mean")
    plt.title(f"Posterior of beta[{j}] ({feature_names[j]})")
    plt.xlabel("beta value")
    plt.ylabel("Density")
    plt.legend()
    plt.tight_layout()
    plt.show()

# ---- tau_j^2 posteriors (local shrinkage) ----
for j in range(p):
    plt.figure()
    plt.hist(tau2_samples[:, j], bins=40, density=True)
    plt.title(f"Posterior of tau^2[{j}] ({feature_names[j]})")
    plt.xlabel("tau^2 value")
    plt.ylabel("Density")
    plt.tight_layout()
    plt.show()

# ---- lambda^2 posterior (global shrinkage) ----
plt.figure()
plt.hist(lambda2_samples, bins=40, density=True)
plt.xlabel(r"$\lambda^2$")
plt.ylabel("Density")
plt.title(r"Posterior of $\lambda^2$")
plt.tight_layout()
plt.show()

# ---- sigma^2 posterior (noise variance) ----
plt.figure()
plt.hist(sigma2_samples, bins=40, density=True)
plt.xlabel(r"$\sigma^2$")
plt.ylabel("Density")
plt.title(r"Posterior of $\sigma^2$")
plt.tight_layout()
plt.show()


# ============================================================
# 6) Posterior predictive at one x0 (Hitters)
# ============================================================

# Pick an example player (centered x0)
x0 = X[0, :]  # 1d array of length p

# Posterior predictive draws: y0^(s) = x0^T beta^(s) + eps^(s)
S = beta_samples.shape[0]
linpred_samples = beta_samples @ x0   # shape (S,)
eps = np.random.normal(scale=np.sqrt(sigma2_samples))  # one eps per draw
y_pred_samples = linpred_samples + eps

# We don't know the true generative model, so just overlay a Gaussian
# using posterior mean and variance as a visual reference
mu_hat = linpred_samples.mean()
sd_hat = np.sqrt(sigma2_samples.mean())

plt.figure()
plt.hist(y_pred_samples, bins=40, density=True, alpha=0.6, label="Posterior predictive")

xs = np.linspace(y_pred_samples.min(), y_pred_samples.max(), 200)
plt.plot(xs, norm.pdf(xs, loc=mu_hat, scale=sd_hat),
         linestyle="--", label="Gaussian approx (mean+var)")

plt.xlabel("log(Salary) at x0 (centered scale)")
plt.ylabel("Density")
plt.title("Posterior predictive at one player (Hitters)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from sklearn.datasets import load_diabetes

# ============================================================
# 1) Inverse-Gaussian sampler  IG(mu, lam)
# ============================================================

def sample_inverse_gaussian(mu, lam):
    """
    Sample from an Inverse-Gaussian IG(mu, lam) using
    the Michael–Schucany–Haas method.
    """
    if mu <= 0 or lam <= 0:
        raise ValueError("mu and lam must be > 0")

    v = np.random.normal()
    y = v**2
    mu2 = mu**2

    x = mu + (mu2 * y) / (2 * lam) - (mu / (2 * lam)) * np.sqrt(
        4 * mu * lam * y + mu2 * y**2
    )
    u = np.random.rand()
    if u <= mu / (mu + x):
        return x
    else:
        return mu2 / x


# ============================================================
# 2) Bayesian lasso Gibbs sampler (centered model)
# ============================================================

def bayesian_lasso_gibbs(
    X,
    y,
    T=6000,
    burn_in=3000,
    a0=1.0,
    b0=1.0,
    r=1.0,
    s=1.0,
    random_state=None,
):
    """
    Fully Bayesian lasso via Gibbs sampling.

    Centered model (no intercept):
        y | beta, sigma^2 ~ N(X beta, sigma^2 I_n)

        beta_j | tau_j^2, sigma^2 ~ N(0, sigma^2 tau_j^2)
        tau_j^2 | lambda^2 ~ Exp(lambda^2 / 2)

        lambda^2 ~ Gamma(r, s)       (shape-rate)
        sigma^2  ~ Inv-Gamma(a0, b0) (shape-scale)

    Assumes y and columns of X are already centered.

    Returns dict of posterior draws for beta, tau^2, lambda^2, sigma^2.
    """
    if random_state is not None:
        np.random.seed(random_state)

    n, p = X.shape

    # Initialize parameters
    beta = np.zeros(p)
    tau2 = np.ones(p)
    lam2 = 1.0
    sigma2 = 1.0

    num_save = T - burn_in
    beta_samples = np.zeros((num_save, p))
    tau2_samples = np.zeros((num_save, p))
    lambda2_samples = np.zeros(num_save)
    sigma2_samples = np.zeros(num_save)

    # Precompute for speed
    XtX = X.T @ X
    Xty = X.T @ y

    for t in range(T):
        # 1) beta | y, tau2, sigma2
        Dtau_inv = np.diag(1.0 / tau2)
        Vbeta_inv = XtX + Dtau_inv
        Vbeta = np.linalg.inv(Vbeta_inv)
        mbeta = Vbeta @ Xty

        # multivariate normal via Cholesky
        L = np.linalg.cholesky(sigma2 * Vbeta)
        z = np.random.normal(size=p)
        beta = mbeta + L @ z

        # 2) tau_j^2 | beta_j, sigma2, lam2 via z_j = 1/tau_j^2
        for j in range(p):
            bj = beta[j] if beta[j] != 0.0 else 1e-8
            mu_j = np.sqrt((lam2 * sigma2) / (bj**2))
            z_j = sample_inverse_gaussian(mu_j, lam2)
            tau2[j] = 1.0 / z_j

        # 3) lambda^2 | tau2
        shape_lam = r + p
        rate_lam = s + 0.5 * tau2.sum()
        lam2 = np.random.gamma(shape_lam, 1.0 / rate_lam)

        # 4) sigma^2 | y, beta, tau2
        resid = y - X @ beta
        SSE = resid @ resid
        quad = np.sum(beta**2 / tau2)

        shape_sig = a0 + 0.5 * (n + p)
        scale_sig = b0 + 0.5 * (SSE + quad)
        sigma2 = 1.0 / np.random.gamma(shape_sig, 1.0 / scale_sig)

        if t >= burn_in:
            idx = t - burn_in
            beta_samples[idx, :] = beta
            tau2_samples[idx, :] = tau2
            lambda2_samples[idx] = lam2
            sigma2_samples[idx] = sigma2

    return {
        "beta": beta_samples,
        "tau2": tau2_samples,
        "lambda2": lambda2_samples,
        "sigma2": sigma2_samples,
    }


# ============================================================
# 3) Load and center diabetes data
# ============================================================

data = load_diabetes()
X_raw = data.data          # (n, p)
y_raw = data.target        # (n,)

n, p = X_raw.shape
print("Diabetes: n, p =", n, p)

# Center X and y
X = X_raw - X_raw.mean(axis=0, keepdims=True)
y = y_raw - y_raw.mean()

# ============================================================
# 4) Run Bayesian lasso on diabetes
# ============================================================

draws = bayesian_lasso_gibbs(
    X,
    y,
    T=6000,
    burn_in=3000,
    a0=1.0,
    b0=1.0,
    r=1.0,
    s=1.0,
    random_state=123,
)

beta_samples   = draws["beta"]
tau2_samples   = draws["tau2"]
lambda2_samples = draws["lambda2"]
sigma2_samples = draws["sigma2"]

S = beta_samples.shape[0]

beta_post_mean = beta_samples.mean(axis=0)
beta_post_sd   = beta_samples.std(axis=0)

print("Posterior mean beta:", np.round(beta_post_mean, 3))
print("Posterior sd beta:  ", np.round(beta_post_sd, 3))

# ============================================================
# 5) Posterior predictive over ALL x, vs empirical y
# ============================================================

# For each posterior draw s, generate y_rep^(s) ~ N(X beta^(s), sigma_s^2 I)
y_pred_all = np.zeros((S, n))

for s in range(S):
    mu_s = X @ beta_samples[s]                       # (n,)
    eps_s = np.random.normal(scale=np.sqrt(sigma2_samples[s]), size=n)
    y_pred_all[s] = mu_s + eps_s                    # (n,)

# Flatten all replicated y's to approximate the marginal predictive of y
y_pred_flat = y_pred_all.ravel()

# Remember: model was fit on centered y; bring both back to original scale
y_obs = y + y_raw.mean()
y_pred = y_pred_flat + y_raw.mean()

plt.figure()
plt.hist(y_obs,  bins=40, density=True, alpha=0.5, label="Observed y (diabetes)")
plt.hist(y_pred, bins=40, density=True, alpha=0.5, label="Posterior predictive")
plt.xlabel("y")
plt.ylabel("Density")
plt.title("Posterior predictive over all x vs empirical distribution of y (diabetes)")
plt.legend()
plt.tight_layout()
plt.show()

# ============================================================
# 6) (Optional) Quick posterior plots for lambda^2 and sigma^2
# ============================================================


plt.figure()
plt.hist(lambda2_samples, bins=40, density=True)
plt.xlabel(r"$\lambda^2$")
plt.ylabel("Density")
plt.title(r"Posterior of $\lambda^2$ (diabetes)")
plt.tight_layout()
plt.show()

plt.figure()
plt.hist(sigma2_samples, bins=40, density=True)
plt.xlabel(r"$\sigma^2$")
plt.ylabel("Density")
plt.title(r"Posterior of $\sigma^2$ (diabetes)")
plt.tight_layout()
plt.show()
```
